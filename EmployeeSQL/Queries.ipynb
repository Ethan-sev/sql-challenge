<span style="color: rgb(43, 43, 43); font-family: Roboto, sans-serif; font-size: 20px; background-color: yellow;">List the first name, last name, and hire date for the employees who were hired in 1986</span>

```sql

SELECT
    last_name AS "Last Name",
    first_name AS "First Name",
    hire_date AS "Hire Date"
    from Employees
    WHERE
    hire_date >= DATE '1986-01-01' AND hire_date < DATE '1987-01-01
LIMIT 10;
```

- <span style="color: var(--vscode-foreground);">List the manager of each department along with their department number, department name, employee number, last name, and first name.</span>

```sql
SELECT
    emp_no AS "Employee Number",
    dept_no as "Dapartment Number",
    (SELECT dept_name FROM Departments WHERE dept_no = Dept_Manager.dept_no) AS "Department Name",
    (select last_name from Employees WHERE emp_no = Dept_Manager.emp_no ),
    (select first_name from Employees WHERE emp_no = Dept_Manager.emp_no )
FROM
    Dept_Manager
    Limit (100);
```

# <span style="color: rgb(43, 43, 43); font-family: Roboto, sans-serif; font-size: 20px; background-color: yellow;">List the department number for each employee along with that employee’s employee number, last name, first name, and department name</span>

```sql
SELECT
    emp_no AS "Employee Number",
    dept_no as "Dapartment Number",
    (SELECT dept_name FROM Departments WHERE dept_no = Dept_Employed.dept_no) AS "Department Name",
    (select last_name from Employees WHERE emp_no = Dept_Employed.emp_no ),
    (select first_name from Employees WHERE emp_no = Dept_Employed.emp_no )
FROM
    Dept_Employed
    Limit (100);
```

1. List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.

```sql
SELECT
    first_name AS "First Name",
    last_name AS "Last Name",
    sex AS "Sex"
FROM
    Employees
WHERE
    first_name = 'Hercules' AND last_name LIKE 'B%'
```
